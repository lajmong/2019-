```python
import pandas as pd

대여이력 = pd.read_csv('대여이력.csv', engine='python', encoding='utf-8')
```

```python
test3 =  pd.read_csv('직선거리.csv', engine='python', encoding='utf-8')
```

```python
test3 = test3.drop('Unnamed: 0', axis =1).drop_duplicates()
```

```python
대여_test = pd.merge(test3, 대여이력, on=['대여대여소명', '반납대여소명'], how = 'inner')
```

```python
대여_test['속력'] = 대여_test.이용거리 / 대여_test.이용시간
```

```python
대여_test=대여_test[대여_test.속력<=600]
대여_test
```

```python
대여_test2=대여_test[['대여대여소명','대여일시','반납대여소명','반납일시','이용시간','이용거리','직선거리','속력']]
```
