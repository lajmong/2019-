```python
import pandas as pd

coor_GU = pd.read_csv('coordinate_구포함.csv')
trans_GU = pd.read_csv('대중교통.csv')
```

```python
from vincenty import vincenty
```

```python
def lat_log(x,y):
    return (x, y)
```

```python
coor_GU.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
coor_GU['vin'] = \
coor_GU.apply(lambda x: lat_log(x.stationLatitude, x.stationLongitude), axis=1)
```

```python
coor_GU = coor_GU[coor_GU.GU != '의정부시']
coor_GU = coor_GU[coor_GU.GU != '광명시']
```

```python
coor_GU
```

```python
gu = list(coor_GU.GU.unique())
gu
```

```python
trans_GU.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
trans_GU['vin'] = trans_GU.apply(lambda x: lat_log(x.YCODE, x.XCODE), axis=1)
```

```python
trans_GU.head()
```

# 마포구

```python
mapo_c = coor_GU[coor_GU.GU == '마포구'].reset_index()
mapo_c.drop('index', axis=1, inplace=True)
mapo_c
```

```python
mapo_t = trans_GU[trans_GU.GU == '마포구'].reset_index()
mapo_t.drop('index', axis=1, inplace=True)
mapo_t
```

```python
mapo_df = pd.DataFrame()
mapo_df['bikeName'] = mapo_c.stationName
mapo_df['transName'] = 0
mapo_df['short'] = None
```

```python
for i in range(len(mapo_c)):
    distance = []
    for j in range(len(mapo_t)):
        distance.append(vincenty(mapo_c.vin[i], mapo_t.vin[j]))
        short = min(distance)
        a = distance.index(short)
        transN = mapo_t.stationName[a]
    mapo_df['transName'][i] = mapo_t.stationName[a]
    mapo_df['short'][i] = short
#mapo_df['transName'] = transName
#mapo_df['short'] = short1
```

```python
mapo_df
```
