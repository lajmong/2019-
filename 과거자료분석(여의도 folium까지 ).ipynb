## Dummy 설정

```python
dummy_y = dummy_y[['stationName', 'label', 'label0', 'label1', 'label2', 'label3', 'stationLatitude', 'stationLongitude']]
```

```python
pivot_dummy = dummy_y.pivot_table(index='stationName', columns=['label0', 'label1', 'label2', 'label3'],
                                  values='label', aggfunc='count').fillna(0)
```
