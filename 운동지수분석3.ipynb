```python
import pandas as pd

대여_test2 = pd.read_csv('대여_test2.csv')
```

```python
대여_test2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
대여_test2
```

#### * 목적지수 : 이용거리 / 직선거리

```python
대여_test2['label'] = 대여_test2.목적지수.apply(lambda x: 0 if x<3\
                                                      else 1)
```

```python
대여_test2.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
type(대여_test2.대여일시)
```

```python
from datetime import datetime
대여_test2['대여일시'] = pd.to_datetime(대여_test2['대여일시'])
```

```python
대여_test2['weekend'] = 0
대여_test2['weekend'] = 대여_test2.대여일시.apply(lambda x : 0 if x.weekday() < 5 else 1)
```

```python
대여_test2['대여시간'] = 대여_test2.대여일시.dt.strftime('%H')
대여_test2['대여시간'] = pd.to_numeric(대여_test2.대여시간)
```

```python
대여_test2['대여시간대'] = 대여_test2.대여시간.apply(lambda x : 0  if (x >1 and x <= 6) \
                                else (1 if (x >6 and x <= 9) \
                                      else(2 if (x >9 and x <= 13) \
                                          else( 3 if (x >13 and x <= 17) \
                                              else( 4 if (x >17 and x <= 20)\
                                                  else 5)))))
```
